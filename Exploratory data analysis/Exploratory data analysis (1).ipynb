```python
#Pandas in python provide an interesting method describe(). 
#The describe function applies basic statistical computations on the dataset like extreme values, count of data points standard deviation etc.
#Any missing value or NaN value is automatically skipped. describe() function gives a good picture of distribution of data.
Df.describe()
```

```python
#value_counts() which can get count of each category in a categorical attributed series of values. 
Df["COTTAGECHEESEQUAN"].value_counts()
```
